Measure Rf and Lf with DC power supply and oscilloscope

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import measureField as mf

#create a 2x3 panels plot
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

tau = np.zeros(6)
meanUSteadyState = np.zeros(6)
meanISteadyState = np.zeros(6)
t0 = np.zeros(6)
t5 = np.zeros(6)
Rf = np.zeros(6)
Lf = np.zeros(6)

#loop for to process each file 
for i in range(2):
    for j in range(3):
        n=3*i+j
        tau[n], meanUSteadyState[n], meanISteadyState[n], t0[n], t5[n], Rf[n], Lf[n] = mf.computeFieldCharacteristic('Data/FieldMeasure_' + str(n+1)+ '.csv', axs[i,j])

fig.tight_layout()

fig, axs = plt.subplots(1, 2, figsize=(15, 10))

axs[0].plot(Rf, '-o')
axs[1].plot(Lf, '-o')

#set labels
axs[0].set_xlabel('Measurement #')
axs[0].set_ylabel('Rf [Ohm]')
axs[1].set_xlabel('Measurement #')
axs[1].set_ylabel('Lf [H]')
axs[0].set_title('Rf')
axs[1].set_title('Lf')

#print mean of Rf and Lf
print('Rf = ', np.mean(Rf), ' +/- ', np.std(Rf))
print('Lf = ', np.mean(Lf), ' +/- ', np.std(Lf))
```
